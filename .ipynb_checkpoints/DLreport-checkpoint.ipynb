The Iraq-Oncology Teaching Hospital/National Center for Cancer Diseases (IQ-OTH/NCCD) lung cancer dataset was collected in the above-mentioned specialist hospitals over a period of three months in fall 2019. It includes CT scans of patients diagnosed with lung cancer in different stages, as well as healthy subjects. IQ-OTH/NCCD slides were marked by oncologists and radiologists in these two centers. The dataset contains a total of 1190 images representing CT scan slices of 110 cases. These cases were grouped into 2 categories according the presence of tumor or not. Initially, the dataset comprised three distinct categories. For the purposes of this study, these categories were consolidated into two broader groups. This adjustment was made to enhance the interpretability of the results and to ensure a more robust analysis by reducing potential class imbalance and increasing the sample size within each group.


**1st part of analysis**

At first, the data set was slit into train and test groups. Then the data were visualized, processed and normalized.

**2nd part of analysis**

A Convulation Neural Network was built. The CNN is comprised of 4 layers and then trained with different epochs. The plots of each training were also produced.

**3rd part of analysis**

The pre-trained Neural Network model EfficientNet was trained with our data. Two trainings took place, the first with 10 epochs and the second with 30 epochs. The results of the trainings were plotted.

**4th part of analysis**

The pre-trained Neural Network model MobileNetV2 was trained with our data. Two trainings took place, the first with 10 epochs and the second with 30 epochs. The results of the trainings were plotted.

**5th part of analysis**

At the end, so as to reach a final conclusion about the best model trained on our data, we compared the accuracy of all three of them and plotted it.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, recall_score, precision_score, classification_report, confusion_matrix
import tensorflow as tf
from PIL import Image
import os
from tensorflow.keras import layers, Sequential
from collections import Counter
from tensorflow import keras
from tensorflow.keras.preprocessing import image_dataset_from_directory
from tensorflow.keras import layers
from tensorflow.keras.layers.experimental import preprocessing
import sklearn.metrics
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.applications import ResNet50
from tensorflow.keras.layers import Dense
from tensorflow.keras.models import Model
from tensorflow.keras.applications import EfficientNetB0
from torchvision.models import efficientnet_b0, EfficientNet_B0_Weights
import torch.optim as optim
import torch
from torchvision import transforms
from torch.utils.data import DataLoader
from tensorflow.keras.callbacks import ModelCheckpoint, EarlyStopping, ReduceLROnPlateau
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D, Dropout

from tensorflow.keras.applications import MobileNetV2
from tensorflow.keras.models import Model
from tensorflow.keras.optimizers import Adam
```

 **Data Loading & Exploratory Data Analysis**

```python
dataset = r'/home/konnie/lungdb'
class_folders = ['Tumor', 'Normal']
target_size = (224, 224) 
valid_extensions = ('.jpg', '.jpeg', '.png')

images = []
labels = []

for class_idx, folder_name in enumerate(class_folders):
    class_path = os.path.join(dataset, folder_name)
    
   
    class_images = [
        os.path.join(class_path, fname)
        for fname in os.listdir(class_path)
        if fname.lower().endswith(valid_extensions)
    ]
  
    for img_path in class_images:
        try:
            with Image.open(img_path) as img:
                if img.mode != 'RGB':
                    img = img.convert('RGB')
                img_resized = img.resize(target_size)
                images.append(np.array(img_resized))
                labels.append(class_idx)
        except Exception as e:
            print(f"Error loading {img_path}: {str(e)}")


images_array = np.array(images)
labels_array = np.array(labels)

print(f"Images array shape: {images_array.shape}")  
print(f"Labels array shape: {labels_array.shape}") 

```

**Visualize the two Classes**

```python
class_names = ['Normal case', 'Tumor case']
valid_extensions = ('.jpg', '.jpeg', '.png', '.bmp', '.tiff')

img_height, img_width = 224, 224
batch_size = 32

train_ds = tf.keras.preprocessing.image_dataset_from_directory(
    dataset,
    labels='inferred',
    label_mode='int', 
    batch_size=batch_size,
    image_size=(img_height, img_width),
    shuffle=True,
    seed=42
)

class_names = train_ds.class_names
print("Class names:", class_names)


plt.figure(figsize=(10, 10))
for images, labels in train_ds.take(1):
    for i in range(9):
        ax = plt.subplot(3, 3, i + 1)
        plt.imshow(images[i].numpy().astype("uint8"))
        plt.title(class_names[labels[i]])
        plt.axis("off")
plt.show()
```

**Split Dataset into Train and Test**

```python
x_train, x_test, y_train, y_test = train_test_split(
    images_array, labels_array, test_size=0.2, random_state=42)

print("The data type of the images is:", x_train.dtype)
print("The shape of the train set:", x_train.shape)
print("The shape of the test set:", x_test.shape)
print("Number of training observations:", x_train.shape[0])
print("Number of testing observations:", x_test.shape[0])
print("Train label distribution:", Counter(y_train))
print("Test label distribution:", Counter(y_test))
```

**Data Preprocessing - Normalization**

```python
print(f"Value interval of the train set before the min-max normalization:[{x_train.min()},{x_train.max()}]")
print(f"Value interval of the test set before the min-max normalization:[{x_test.min()},{x_test.max()}]")
x_train = x_train / 255.0
x_test = x_test / 255.0
print(f"Value interval of the train set after the min-max normalization:[{x_train.min()},{x_train.max()}]")
print(f"Value interval of the test set after the min-max normalization:[{x_test.min()},{x_test.max()}]")
```

**CNN Overview**

```python
cnn_model = tf.keras.models.Sequential([
    tf.keras.Input(shape=(224, 224, 3)),         
                    
    layers.Conv2D(32, (3,3), padding="same", activation="relu"),
    layers.MaxPooling2D((2,2)),

    layers.Conv2D(64, (3,3), padding="same", activation="relu"),
    layers.MaxPooling2D((2,2)),

    layers.Conv2D(128, (3,3), padding="same", activation="relu"),
    layers.MaxPooling2D((2,2)),

    layers.Conv2D(256, (3,3), padding="same", activation="relu"),
    layers.MaxPooling2D((2,2)),

    layers.Flatten(),
    layers.Dropout(0.5),
    layers.Dense(512, activation="relu"),
    layers.Dense(2, activation="softmax") 
])

cnn_model.summary()
```

**Data Augmentation using ImageDataGenerator**

-Prevents overfitting

-Increases training set size

-Improves model generalization

```python
data_augmentation = ImageDataGenerator(
    rotation_range=15,      
    width_shift_range=0.1,  
    height_shift_range=0.1, 
    horizontal_flip=True,   
    zoom_range=0.1,         
    fill_mode='nearest',    
    validation_split=0.2    
)
```

**Compile the model without augmentation**

```python
cnn_model.compile(loss="sparse_categorical_crossentropy", optimizer="adam", metrics=["accuracy"])
```

**Train and Validation Generator**

```python
train_generator = data_augmentation.flow(
    x_train, y_train,
    batch_size=32,
    subset='training'
)

validation_generator = data_augmentation.flow(
    x_train, y_train, 
    batch_size=32,
    subset='validation',
    shuffle=False  
)
```

**Fitting the model**

```python
history_cnn1 = cnn_model.fit(
    train_generator,
    steps_per_epoch=len(x_train) * 0.8 // 32,  # 80% of data is used for training with validation_split=0.2
    epochs=10,
    validation_data=validation_generator,
    validation_steps=len(x_train) * 0.2 // 32,  # 20% of data is used for validation
)
```

**Accuracy and Loss plots**

```python
def plot_training_history(history, title):
    """Plot the training and validation accuracy/loss"""
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5))
    
    # Accuracy
    ax1.plot(history.history['accuracy'])
    ax1.plot(history.history['val_accuracy'])
    ax1.set_title(f'{title} - Accuracy')
    ax1.set_ylabel('Accuracy')
    ax1.set_xlabel('Epoch')
    ax1.legend(['Train', 'Validation'], loc='lower right')
    
    # Loss
    ax2.plot(history.history['loss'])
    ax2.plot(history.history['val_loss'])
    ax2.set_title(f'{title} - Loss')
    ax2.set_ylabel('Loss')
    ax2.set_xlabel('Epoch')
    ax2.legend(['Train', 'Validation'], loc='upper right')
    
    plt.tight_layout()
    plt.show()
```

```python
plot_training_history(history_cnn1, "CNN with 10 epochs")
```

**Train the CNN with 30 epochs and plot accuracy and loss**

```python
history_cnn2 = cnn_model.fit(
    train_generator,
    steps_per_epoch=len(x_train) * 0.8 // 32,
    epochs=30,
    validation_data=validation_generator,
    validation_steps=len(x_train) * 0.2 // 32,
)
```

```python
plot_training_history(history_cnn2, "Simple CNN")
```

**Train the CNN with 50 epochs and plot the accuracy and loss**

```python
history_cnn3 = cnn_model.fit(
    train_generator,
    steps_per_epoch=len(x_train) * 0.8 // 32,
    epochs= 50,
    validation_data=validation_generator,
    validation_steps=len(x_train) * 0.2 // 32,
)
```

```python
plot_training_history(history_cnn3, "Simple CNN")
```

**Compare the accuracy of the trainings**

```python
import matplotlib.pyplot as plt
import numpy as np

def plot_accuracy_comparison(histories, names=None, figsize=(10, 6)):

    if names is None:
        names = [f"Model {i+1}" for i in range(len(histories))]
        
    plt.figure(figsize=figsize)
    
    colors = ['blue', 'orange', 'green']  
    line_styles = ['-', '--']  
    
    for i, history in enumerate(histories):
        if hasattr(history, 'history'):
            history = history.history
            
        # Plot training accuracy
        epochs = range(1, len(history['accuracy']) + 1)
        plt.plot(epochs, history['accuracy'], 
                 color=colors[i], linestyle=line_styles[0],
                 marker='o', markersize=4, 
                 label=f'{names[i]} (train)')
        
        # Plot validation accuracy
        if 'val_accuracy' in history:
            plt.plot(epochs, history['val_accuracy'], 
                     color=colors[i], linestyle=line_styles[1],
                     marker='s', markersize=4, 
                     label=f'{names[i]} (val)')
    
    plt.title('Model Accuracy Comparison', fontsize=16)
    plt.xlabel('Epochs', fontsize=14)
    plt.ylabel('Accuracy', fontsize=14)
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.legend(loc='lower right')
    
  
    y_vals = []
    for h in histories:
        if hasattr(h, 'history'):
            y_vals.extend(h.history['accuracy'])
        else:
            y_vals.extend(h['accuracy'])
    
    y_min = max(0, min(y_vals) - 0.15)  # Leave some margin
    plt.ylim(bottom=y_min, top=1.0)
    
    plt.tight_layout()
    plt.show()

    


```

**Plots comparing the training runs**


```python
plot_accuracy_comparison(
    [history_cnn1, history_cnn2, history_cnn3], 
    names=['First Run', 'Second Run', 'Third Run']
)
```


The model was trained for 10, 30 and 50 epochs, and the results were evaluated using accuracy and loss plots for training and validation data. These key observations were made: 

First Run (10 epochs, blue): Starts with lower accuracy (~0.6) and shows significant fluctuations, with both training and validation accuracy staying below 0.75.

Second Run (30 epochs, orange): Begins at a higher accuracy (~0.7) and demonstrates improved stability and performance, with both metrics rising to around 0.8 by the end of training.

Third Run (50 epochs, green): Achieves the best results, with training and validation accuracy starting above 0.8 and consistently reaching and maintaining values above 0.85, peaking near 0.9. The curves are also more stable, indicating better model generalization and less overfitting.

Overall, the plot demonstrates clear progressive improvement in model performance and stability across the three runs, with the Third Run achieving the highest and most consistent accuracy for both training and validation sets.


**Evaluation of models**

```python
def evaluate_model(model, test_generator, save_results=True):
    print("Evaluating model performance...")
    
    # Run evaluation on test/validation set
    results = model.evaluate(test_generator, verbose=1)
    
    # Extract metrics
    loss = results[0]
    accuracy = results[1]
    
 
    metrics = {
        'loss': loss,
        'accuracy': accuracy
    }
    

    metrics_names = model.metrics_names
    for i, metric_name in enumerate(metrics_names):
        if metric_name not in ['loss', 'accuracy']:
            metrics[metric_name] = results[i]
    
    print("\n" + "="*50)
    print("FINAL MODEL EVALUATION")
    print("="*50)
    print(f"Loss:     {loss:.4f}")
    print(f"Accuracy: {accuracy:.4f} ({accuracy*100:.2f}%)")
    
    for metric_name, value in metrics.items():
        if metric_name not in ['loss', 'accuracy']:
            print(f"{metric_name}: {value:.4f}")
    print("="*50)
    
    return metrics
```

**Evaluation of CNN**

```python
metrics = evaluate_model(cnn_model, validation_generator)

final_accuracy = metrics['accuracy']
final_loss = metrics['loss']

print(f"Final model accuracy: {final_accuracy:.4f}")
print(f"Final model loss: {final_loss:.4f}")
```

**Efficient Net**

```python
IMG_HEIGHT = 224
IMG_WIDTH = 224

base_model = EfficientNetB0(weights='imagenet', 
                           include_top=False, 
                           input_shape=(IMG_HEIGHT, IMG_WIDTH, 3))


base_model.trainable = False

x = base_model.output
x = GlobalAveragePooling2D()(x)
x = Dropout(0.5)(x)  
x = Dense(128, activation='relu')(x)
x = Dropout(0.3)(x)  
num_classes = 2  
predictions = Dense(1, activation='softmax')(x)


model_eff = Model(inputs=base_model.input, outputs=predictions)
```

**Compile the model**

```python
model_eff.compile(
    optimizer= 'adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)
```

**Early stopping & reduce learning rate**

```python
early_stopping = EarlyStopping(
    monitor='val_loss',
    patience=10,
    restore_best_weights=True,
    verbose=1
)

reduce_lr = ReduceLROnPlateau(
    monitor='val_loss',
    factor=0.2,
    patience=5,
    min_lr=0.00001,
    verbose=1
)
```

**Train the model**

```python
callbacks = [early_stopping, reduce_lr]

history_eff10 = model_eff.fit(
    train_generator,
    epochs= 10,
    validation_data=validation_generator,
    callbacks = callbacks
    
)
```

**Plot the accuracy and loss of EfficientNet**

```python
plot_training_history(history_eff10, "Efficient net")
```

**Train EfficientNet with 30 epochs and plot the accuracy and loss**

```python
history_eff30 = model_eff.fit(
    train_generator,
    epochs= 30,
    validation_data=validation_generator,
    callbacks = callbacks
    
)
```

```python
plot_training_history(history_eff30, "Efficient net")
```

**Evaluate EfficientNet**

```python
metrics = evaluate_model(model_eff, validation_generator)
final_accuracy = metrics['accuracy']
final_loss = metrics['loss']

print(f"Final model accuracy: {final_accuracy:.4f}")
print(f"Final model loss: {final_loss:.4f}")
```

```python
plot_accuracy_comparison(
    [history_eff10, history_eff30], 
    names=['First Run', 'Second Run']
)
```

This plot displays the training and validation accuracy across epochs for two runs of a model. Both training and validation accuracy curves for each run are nearly flat, with the "First Run" maintaining a training accuracy around 0.4 and validation accuracy around 0.3, and the "Second Run" showing almost identical values. There is minimal to no improvement in accuracy over the course of training for either run, indicating that the model did not learn effectively during these runs. This could suggest issues such as underfitting or  poor data preprocessing.

**MobileNetV2**

**Load the model**

```python
IMG_HEIGHT = 224
IMG_WIDTH = 224

base_model_mobnet = MobileNetV2(
    weights='imagenet',
    include_top=False,
    input_shape=(IMG_HEIGHT, IMG_WIDTH, 3)
)

base_model_mobnet.trainable = False
```

**Build the model**

```python
x = base_model_mobnet.output
x = GlobalAveragePooling2D()(x)
x = Dropout(0.5)(x)
x = Dense(128, activation='relu')(x)
x = Dropout(0.3)(x)


predictions = Dense(1, activation='sigmoid')(x)
model_mobnet = Model(inputs=base_model_mobnet.input, outputs=predictions)
```

**Compile model**

```python
model_mobnet.compile(
    optimizer=Adam(learning_rate=0.001),
    loss='binary_crossentropy',
    metrics=['accuracy']
)

model_mobnet.summary()
```

**Add early stopping and reduction of learning rate**

```python
early_stopping = EarlyStopping(
    monitor='val_loss',
    patience=10,
    restore_best_weights=True,
    verbose=1
)

reduce_lr = ReduceLROnPlateau(
    monitor='val_loss',
    factor=0.2,
    patience=3,
    min_lr=0.00001,
    verbose=1
)

callbacks = [early_stopping, reduce_lr]
```

**Train model**

```python
# Train the model
history_mobnet10 = model_mobnet.fit(
    train_generator,
    epochs=10,
    validation_data=validation_generator,
    callbacks=callbacks
)
```

**Plot the accuracy and the loss**

```python
plot_training_history(history_mobnet10, "Mobile Net")
```

**Train again with 30 epochs**

```python
history_mobnet30 = model_mobnet.fit(
    train_generator,
    epochs=30,
    validation_data=validation_generator,
    callbacks=callbacks
)
```

```python
plot_training_history(history_mobnet30, "Mobile Net")
```

**Evaluate the model**

```python
metrics = evaluate_model(model_mobnet, validation_generator)

final_accuracy = metrics['accuracy']
final_loss = metrics['loss']

print(f"Final model accuracy: {final_accuracy:.4f}")
print(f"Final model loss: {final_loss:.4f}")
```

```python
plot_accuracy_comparison(
    [history_mobnet10, history_mobnet30], 
    names=['First Run', 'Second Run', 'Third Run']
)
```

First Run (10 epochs, blue): Starts with a lower training accuracy (around 0.75) but shows steady improvement, reaching above 0.9 by the end of training. The validation accuracy for this run is higher than the training accuracy and remains relatively stable, almost reaching 0.95.

Second Run (30 epochs, orange): Both training and validation accuracy start high (around 0.9) and maintain this level throughout the epochs, with minor fluctuations. The validation accuracy is close to the training accuracy, indicating good generalization and stability.

```python
import pandas as pd

def model_summary(history_cnn2, history_eff30, history_mobnet30):
    # Extract final epoch values for each model
    models_data = {
        'CNN': {
            'Train Accuracy': history_cnn2.history['accuracy'][-1],
            'Val Accuracy': history_cnn2.history['val_accuracy'][-1],
            'Train Loss': history_cnn2.history['loss'][-1],
            'Val Loss': history_cnn2.history['val_loss'][-1]
        },
        'EfficientNet': {
            'Train Accuracy': history_eff30.history['accuracy'][-1],
            'Val Accuracy': history_eff30.history['val_accuracy'][-1],
            'Train Loss': history_eff30.history['loss'][-1],
            'Val Loss': history_eff30.history['val_loss'][-1]
        },
        'MobileNetV2': {
            'Train Accuracy': history_mobnet30.history['accuracy'][-1],
            'Val Accuracy': history_mobnet30.history['val_accuracy'][-1],
            'Train Loss': history_mobnet30.history['loss'][-1],
            'Val Loss': history_mobnet30.history['val_loss'][-1]
        }
    }
    
    # Create DataFrame
    summary_df = pd.DataFrame(models_data).T
    
    # Round values for better readability
    summary_df = summary_df.round(4)
    
    return summary_df


summary =model_summary(history_cnn2, history_eff30, history_mobnet30)
print("Model Performance Summary:")
print(summary)
```

**MobileNetV2 (Green):**
Both training and validation accuracy for MobileNetV2 are consistently high, staying above 0.9. The curves are stable and closely aligned, indicating excellent generalization and minimal overfitting.

**Custom CNN (Blue):**
The custom CNN shows moderate performance, with training and validation accuracy fluctuating between 0.7 and 0.8. The curves are more variable, suggesting less stability and possibly sensitivity to data or training conditions.

**EfficientNet (Orange):**
EfficientNet performs poorly in this experiment, with both training and validation accuracy remaining flat and low (around 0.4 for training and 0.3 for validation), indicating that the model failed to learn meaningful patterns from the data.

The significant difference performance-wise between the two pre-trained models is likely due to MobileNetV2’s efficiency, suitability for smaller datasets, and robustness to limited data, whereas EfficientNet’s complexity and data requirements were not matched by this experimental setup.

```python
def predict_and_display(model, image_generator, num_images=5):
    """Make predictions and display results"""
    images, true_labels = next(image_generator)
    predictions = model.predict(images)
    
    plt.figure(figsize=(15, 10))
    
    for i in range(num_images):
        plt.subplot(1, num_images, i+1)
        plt.imshow(images[i])
        predicted_class = "Dog" if predictions[i][0] > 0.5 else "Cat"
        true_class = "Dog" if true_labels[i] > 0.5 else "Cat"
        title = f"Pred: {predicted_class}\nTrue: {true_class}"
        plt.title(title, color=("green" if predicted_class == true_class else "red"))
        plt.axis('off')
    
    plt.tight_layout()
    plt.show()
```
